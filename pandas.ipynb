```python
import numpy as np
import pandas as pd

df = pd.read_csv('movies_2.csv') # Gathering data
df.head(5)
```

```python
#Assessing data

#shape of Dataframe
df.shape

#All info of Dataframe
df.info()

# just the types of each column (again)
df.dtypes

# NAN values that look like Boolean mask
df.isna()
```

```python
#drop last row(containing all NAN)
df=df.drop(labels=615,axis=0)

#drop summary column
df=df.drop(labels='Summary',axis=1)


```

```python
#rename method to change all columns names lowercase and add an underscore if they are made of 2 words

column_mapper={ 'MovieID':'movie_id',
                'Title':'title', 
            	'MPAA Rating':'mpaa_rating',
                'Budget':'budget',
                'Gross':'gross',
            	'Release Date':'release_date',
            	'Genre':'genre'	,
                'Runtime':'runtime',
                'Rating':'rating',
                'Rating Count':'rating_count',
               }

df=df.rename(mapper=column_mapper, axis=1)

df
```

```python
#use fillna() method to replace the missing values of Rating with their median
df["rating"].fillna(value=df["rating"].median(), inplace=True)
df
```

```python
#The Rating Count column is also skewed, use the same method to fill the missing values.
df['rating_count'].fillna(value=df['rating_count'].median(), inplace=True)
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.info()
```

```python
#Convert Budget into integer type (use astype() method)
#Convert Gross into integer type
df=df.astype({'budget':'int','gross':'int64' })

#Convert Release Date into date type (use to_datetime method)
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d')

df.dtypes
```

```python
#  show the movies with more than 7 in Rating & greater than 50 million Gross
rating_7_plus=df['rating']>7
gross_50m_plus=df['gross']>50000000
print(df[(rating_7_plus)&(gross_50m_plus)])
```

```python
# show the movies with more than 7 in Rating & greater than 50 million Gross & 
#with Parental guidance as MPAA Rating


mpaa_pg=df['mpaa_rating']=='PG'
print(df[(rating_7_plus)&(gross_50m_plus)&(mpaa_pg)])
```

```python
#count of Animation movies with more than 7 in Rating (use the shape() function)

is_Animation= df['genre']=='Animation'
to_count= df[(rating_7_plus)&(is_Animation)]
to_count.shape

#Answer: 39
```

```python
#show the list of top 5 movies based on Budget
df.sort_values(by='budget', ascending=False).head()
```

```python
# show the top 5 Comedy movies by Rating

is_Comedy= df['genre']=='Comedy'

df[is_Comedy].sort_values(by='rating', ascending=False).head()
```

```python
#top 5 movie names by Rating

df.sort_values(by='rating', ascending=False).head()['title']

```

```python
#top 3 high Gross Romance movies released after 1999, not included (typecast it to datetime)



is_Romance= df['genre']=='Romance'
df['year'] = pd.DatetimeIndex(df['release_date']).year
after_1999 = df['year']>1999
df[(is_Romance)&(after_1999)].sort_values(by='budget', ascending=False).head(3)

```

```python
#   how many Genres are present in the dataframe? (use the function value_counts()

print('nr of different genre:',df['genre'].value_counts().count())

df['genre'].value_counts()
```

```python
#top 5 expensive movies released after 1999, not included (measured by Budget)

df[after_1999].sort_values(by='budget', ascending=False).head()
```

```python
#  most & least frequent MPAA Rating in the dataset in terms of occurrences

mpaa_counts=df['mpaa_rating'].value_counts()
print('most frequent mpaa:', mpaa_counts.index[0])
print('least frequent mpaa:', mpaa_counts.index[-1])

```

```python
# most and least expensive Genre 
#(take an average of all Budget measures grouped by Genre - use groupBy() method)
genre_cost=df.groupby('genre')['budget'].mean()
genre_cost.sort_values(ascending=False)
```

```python
# which Genre is favored the most by the people?
df.groupby('genre')['rating'].mean().sort_values(ascending=False)

```
